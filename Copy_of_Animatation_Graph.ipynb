```python
# import bar_chart_race as bcr

# fig, ax = plt.subplots(1, 1, figsize=(5, 3), dpi=120, tight_layout=True)

# bcr.bar_chart_race(df=df, filename=None, fig = fig, title = 'Animation Chart', orientation = 'v', fixed_order =True)
# # plt.xticks(np.arange(min(year), max(year) + 1, 5.0))
# ax.set_ylim(ymin=15)
# colss= ['Mean Skin Temperature (°C)','Head','Face','Neck','Left Foot','Lower Left Leg','Upper Left Leg','Abdomen','Left Shoulder',
#         'Right Shoulder','Thorax','Left Hand', 'Lower Left Arm', 'Upper Left Arm', 'Lower Right Arm', 'Right Arm',
#         'Upper Right Arm', 'Lower Right Leg', 'Right Foot', 'Upper Right Leg']
# colss2= ['Head','Face','Neck','Left Foot','Lower Left Leg','Upper Left Leg','Abdomen','Left Shoulder',
#         'Right Shoulder','Thorax','Left Hand', 'Lower Left Arm', 'Upper Left Arm', 'Lower Right Arm', 'Right Arm',
#         'Upper Right Arm', 'Lower Right Leg', 'Right Foot', 'Upper Right Leg']

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots




# px.line(df1, x='Date', y='value', color='AAPL' )


fig1 = px.bar(df_melt, x='Body Segments', y= 'Temprature', animation_frame="time (sec)",
              color="Body Segments")
# fig1.update_layout(barmode='group')


fig2 = px.line(
    df_melt1,
    x='variable',
    y= 'value',  # 2nd dataframe,
    color = "color",
    animation_frame="time (sec)",
    color_discrete_sequence=["#16537e"] + px.colors.qualitative.Alphabet
)
fig2.data[0].name="Cold but Comfortable"
fig2.update_traces(line_color='#16537e', showlegend=True)

fig3 = px.line(
    df_melt2,
    x='variable',
    y= 'value',  # 2nd dataframe
    color = "color",
    animation_frame="time (sec)",
    color_discrete_sequence=["#cc0000"] + px.colors.qualitative.Alphabet
)
fig3.data[0].name="Warm but Comfortable"
fig3.update_traces(line_color='#cc0000', showlegend=True)

fig4 = px.line(
    df_melt3,
    x='variable',
    y= 'value',  # 2nd dataframe
    color = "color",
    animation_frame="time (sec)",
    color_discrete_sequence=["#6aa84f"] + px.colors.qualitative.Alphabet
)
fig4.data[0].name="Comfortable"
fig4.update_traces(line_color='#6aa84f', showlegend=True)


fig5 = go.Figure(
    data=fig1.data + fig2.data  + fig3.data + fig4.data,
    frames=[
        go.Frame(data=fr1.data + fr2.data + fr3.data + fr4.data, name=fr1.name)
        for fr1, fr2, fr3, fr4 in zip(fig1.frames, fig2.frames, fig3.frames, fig4.frames)
    ],
    layout=fig1.layout,
)

fig5.layout.update(fig1.layout)
fig5.layout.update(fig2.layout)
fig5.layout.update(fig3.layout)
fig5.layout.update(fig4.layout)

fig5.update_layout(margin=dict(l=30, r=30, t=50, b=200),
                   paper_bgcolor="LightSteelBlue", xaxis_title="Body Segments",
                   yaxis_title="Temprature(°C)", legend_title_text='Legend', title='MRT = +40deg', yaxis_range=[0,35])
fig5['layout']['updatemenus'][0]['pad']=dict(r= 10, t= 150)
fig5['layout']['sliders'][0]['pad']=dict(r= 10, t= 150)

fig5



# trace1 = px.bar(
#     df1,
#     x=px.Constant(' '),
#     y= colss,
#     animation_frame="time"
# )

# trace2 = go.Scatter(
#     x=df2['Body Segments'], y=df2.columns[1:4], yaxis='y2'
# )

# fig = make_subplots(specs=[[{"secondary_y": True}]])
# fig.add_trace(trace1).data[0]
# fig.add_trace(trace2,secondary_y=True)
# fig['layout'].update_layout(xaxis_title="Body Segments", yaxis_title="Temprature(°C)", legend_title_text='Body Segments', barmode='group')
# fig



# fig3
# fig.show()

# fig5.write_html("+40deg.html")

# import matplotlib.animation as animation
# from matplotlib.animation import FuncAnimation
# from matplotlib import pyplot as plt
```
